```python
# Последовательная модель НС
from tensorflow.keras.models import Sequential

# Основные слои
from tensorflow.keras.layers import Dense, Activation, Dropout, BatchNormalization

# Утилиты для to_categorical()
from tensorflow.keras import utils

# Алгоритмы оптимизации для обучения модели
from tensorflow.keras.optimizers import Adam, Adadelta

# Библиотека для работы с массивами
import numpy as np

# Библиотека для работы с таблицами
import pandas as pd

# Отрисовка графиков
import matplotlib.pyplot as plt

# Связь с google-диском
from google.colab import files

# Предварительная обработка данных
from sklearn import preprocessing

# Разделение данных на выборки
from sklearn.model_selection import train_test_split

# Для загрузки датасета
from keras.datasets import fashion_mnist

# Отрисовывать изображения в ноутбуке, а не в консоль или файл
%matplotlib inline
```

##Описание базы

###База: одежда, обувь и аксессуары
- Датасет состоит из набора изображений одежды, обуви, аксессуаров и их классов.
- Изображения одного вида хранятся в numpy-массиве **(28, 28)** - **x_train, x_test**.
- База содержит **10** классов: (Футболка, Брюки, Пуловер, Платье, Пальто, Сандалии/Босоножки, Рубашка, Кроссовки, Сумочка, Ботильоны) - **y_train, y_test**.
- Примеров: train - **60000**, test - **10000**.

###Вывод примеров

```python
# Загрузка датасета
(x_train_org, y_train_org), (x_test, y_test) = fashion_mnist.load_data()

# Вывод размерностей выборок
x_train, x_val = x_train_org[:50000], x_train_org[50000:]
y_train, y_val = y_train_org[:50000], y_train_org[50000:]

print(x_train.shape)
print(x_val.shape)
print(x_test.shape)
```

```python
x_train = x_train.astype('float32') / 255.
x_val = x_val.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.
```

```python
x_train = x_train.reshape(x_train.shape[0], -1)
x_val   = x_val.reshape(x_val.shape[0], -1)
x_test  = x_test.reshape(x_test.shape[0], -1)
```

```python
CLASS_COUNT = 10
```

```python
y_train = utils.to_categorical(y_train, CLASS_COUNT)
y_val = utils.to_categorical(y_val, CLASS_COUNT)
y_test = utils.to_categorical(y_test, CLASS_COUNT)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_val.shape)
print(y_val.shape)
```

```python
def create_model(units1, units2, activation1, activation2):
    model = Sequential()

    model.add(Dense(units1, input_dim=x_train.shape[1], activation=activation1))

    model.add(Dense(units2, activation=activation2))
    model.add(Dense(CLASS_COUNT, activation='softmax'))

    model.compile(
        loss='categorical_crossentropy',
        optimizer='adam',
        metrics=['accuracy']
    )

    return model
```

```python
val_accuracies = {}
test_accuracies = {}
```

```python
model1 = create_model(60, 30, 'relu', 'relu')
history1 = model1.fit(x_train,        # обучающая выборка, входные данные
          y_train,        # обучающая выборка, выходные данные
          batch_size=64,# кол-во примеров, которое обрабатывает нейронка перед одним изменением весов
          validation_data=(x_val, y_val),
          epochs=5,      # количество эпох, когда нейронка обучается на всех примерах выборки
          verbose=1)

val_accuracies['model1'] = history1.history['val_accuracy'][-1]  # последняя эпоха
test_loss, test_acc = model1.evaluate(x_test, y_test, verbose=0)
print()
test_accuracies['model1'] = test_acc
print("Test accuracy:", test_accuracies['model1'])
```

Модели с только tanh или sigmoid показывают немного меньшую точность и более медленное обучение.
